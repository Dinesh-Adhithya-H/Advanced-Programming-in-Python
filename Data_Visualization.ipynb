```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
import random
import collections
import statsmodels.api as sm
```

# Histogram

```python
np.random.seed(10)
ages=np.random.normal(28,5,30)
plt.hist(ages,bins=10)
plt.xlabel("Age")
plt.ylabel("Frequency")
```

# Density plot

```python
np.random.seed(10)
x_values=np.random.normal(8,5,50)
df=pd.DataFrame(x_values,columns=["age"])
df.plot(kind="density")
```

# Bar plot

```python
sample_colors=["red","green","blue","black"]
ball_colors=np.random.choice(sample_colors,100)
ball_colors_count=collections.Counter(ball_colors)
colors=list(ball_colors_count.keys())
values=list(ball_colors_count.values())

plt.bar(colors,values,color="blue",width=0.5)
```

# Box plot for 1 variable

```python
np.random.seed(99)
ages=np.random.normal(28,5,30)
plt.boxplot(ages)
plt.ylabel("age")
```

# Box plot in 2 variable

```python
np.random.seed(9999)
ages = np.random.normal(60, 5, 30)
weights = np.random.normal(60, 5, 30)
df =pd.DataFrame({'ages':ages, 'weights': weights})
boxplot = df.boxplot()
```

# QQ plots

```python
random_normals=[np.random.normal(1000) for i in np.arange(1000)]
```

```python
sm.qqplot(np.array(random_normals),line="45")
```

```python
random_uniform = [random.random() for i in range(1000)]
sm.qqplot(np.array(random_uniform), line='45')
plt.show()
```

# Scatter plot

```python
h=np.random.normal(5,1,100)
w=np.random.normal(120,150,100)
```

```python
plt.scatter(h,w)
```

```python
df=pd.DataFrame({'h':h,"w":w})
```

```python
df.plot.scatter(x="h",y="w")
```

# scatter plot matrix

```python
A = np.random.normal(0,1,50)
B = np.random.normal(10,2,50)
C = np.random.uniform(30,15,50)
D = np.random.uniform(4,1,50)
df = pd.DataFrame({'A': A, 'B': B,'C': C, 'D': D})
pd.plotting.scatter_matrix(df, diagonal='kde')
plt.show()
```
