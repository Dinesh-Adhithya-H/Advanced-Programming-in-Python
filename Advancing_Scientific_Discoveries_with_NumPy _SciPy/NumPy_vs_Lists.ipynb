# Memory

```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import time
```

```python
s=range(1000)
sys.getsizeof(s)*len(s)
```

```python
s1=np.arange(1000)
s1.itemsize*s1.size
```

```python
x=[]
y1=[]
y2=[]
for j in range(100000):
  s=range(j)
  s1=np.arange(j)
  x.append(j)
  y1.append(sys.getsizeof(s)*len(s))
  y2.append(s1.itemsize*s1.size)
```

```python
plt.plot(x,y1,label="Lists")
plt.plot(x,y2,label="NumPy")
plt.xlabel("No. of elements in array")
plt.ylabel("Memory (in bytes)")
plt.legend()
plt.show()
```

## Sum operation

```python
x=[]
y1=[]
y2=[]
for j in range(10000,20000):
  
  s1=range(j)
  s2=range(j)

  t1=time.time()

  s3=[i+j for (i,j) in zip(s1,s2)]

  t2=time.time()

  y1.append(t2-t1)

  s4=np.arange(j)
  s5=np.arange(j)

  t1=time.time()

  
  s6=s4+s5

  t2=time.time()

  y2.append(t2-t1)
```

```python
plt.plot(x,y1,label="Lists")
plt.plot(x,y2,label="NumPy")
plt.xlabel("No. of elements in array")
plt.ylabel("Time (in seconds)")
plt.legend()
plt.show()
```

## Multiplication

```python
x=[]
y1=[]
y2=[]
for j in range(10000,20000):
  
  s1=range(j)
  s2=range(j)

  t1=time.time()

  s3=[i*j for (i,j) in zip(s1,s2)]

  t2=time.time()

  y1.append(t2-t1)

  s4=np.arange(j)
  s5=np.arange(j)

  t1=time.time()
  
  s6=s4*s5

  t2=time.time()
  
  y2.append(t2-t1)
```

```python
x=np.arange(10000,20000)
plt.plot(x,y1,label="Lists")
plt.plot(x,y2,label="NumPy")
plt.xlabel("No. of elements in array")
plt.ylabel("Time (in seconds)")
plt.legend()
plt.show()
```

## Appending

```python
x=[]
y1=[]
y2=[]
for j in range(10000,20000):
  
  s1=list(range(j))

  t1=time.time()

  s1.append(0)

  t2=time.time()

  y1.append(t2-t1)

  s4=np.arange(j)

  t1=time.time()

  np.append(s4,0)

  t2=time.time()
  y2.append(t2-t1)
```

```python
x=np.arange(10000,20000)
plt.plot(x,y1,label="Lists")
plt.plot(x,y2,label="NumPy")
plt.xlabel("No. of elements in array")
plt.ylabel("Time (in seconds)")
plt.legend()
plt.show()
```

## Searching

```python
np.arange(5)
```

```python
x=[]
y1=[]
y2=[]
for j in range(10000,20000):
  
  s1=list(range(j))

  t1=time.time()

  s1.index(j-1)

  t2=time.time()

  y1.append(t2-t1)

  s4=np.arange(j)

  t1=time.time()
  
  np.where(s4==j-1)

  t2=time.time()
  y2.append(t2-t1)
```

```python
x=np.arange(10000,20000)
plt.plot(x,y1,label="Lists")
plt.plot(x,y2,label="NumPy")
plt.xlabel("No. of elements in array")
plt.ylabel("Time (in seconds)")
plt.legend()
plt.show()
```

## Deletion

```python
x=[]
y1=[]
y2=[]
for j in range(10000,20000):
  
  s1=list(range(j))

  t1=time.time()

  s1.remove(j-1)

  t2=time.time()

  y1.append(t2-t1)

  s4=np.arange(j)

  t1=time.time()

 

  np.delete(s4,j-1)

  t2=time.time()
  y2.append(t2-t1)
```

```python
x=np.arange(10000,20000)
plt.plot(x,y1,label="Lists")
plt.plot(x,y2,label="NumPy")
plt.xlabel("No. of elements in array")
plt.ylabel("Time (in seconds)")
plt.legend()
plt.show()
```

## Updation

```python
x=[]
y1=[]
y2=[]
for j in range(10000,20000):
  
  s1=list(range(j))

  t1=time.time()

  s1[j-1]=0

  t2=time.time()

  y1.append(t2-t1)

  s4=np.arange(j)

  t1=time.time()

  s4[j-1]=0

  t2=time.time()

  y2.append(t2-t1)
```

```python
x=np.arange(10000,20000)
plt.plot(x,y1,label="Lists")
plt.plot(x,y2,label="NumPy")
plt.xlabel("No. of elements in array")
plt.ylabel("Time (in seconds)")
plt.legend()
plt.show()
```
