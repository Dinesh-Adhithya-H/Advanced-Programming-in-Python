# Exercise 3: Pig Latin

Pig Latin is a children's secret language. To translate a word into Pig Latin, follow these rules:

1. If the first letter is a vowel (a, e, i, o, or u) then add `way` to the end of the word.
2. Otherwise, move the first letter to the end of the word, and add `ay`.

Some examples:

- `computer` -> `omputercay` 
- `bhopal` -> `hopalbay`
- `table` -> `abletay`
- `apple` -> `appleway`
- `elephant` -> `elephantway`
- `iiser` -> `iiserway`
- `papaya` -> `apayapay`


1. Ask the user to enter a word in English, all lowercase, no punctuation.
2. Print the translation of this word into Pig Latin.

````python
vowels=["a","e","i","o","u"]
word=str(input("Enter a word (all lowercase, no punctuation) :"))
if word[0] in vowels:
    print(word+"way")
else:
    print(word[1:]+word[0]+"ay")
````

    - If they enter something that is *not* on the menu, then say that you don't have that.
    
Example:

    Order: sandwich

### Hints/notes    

- Define the `menu` dict at the top with keys and values

````python
dict={"sandwich":10,"water":3}
tot=0
while True:
    order=str(input("Order:"))
    if order=="":
        print("your total is:",tot)
        break
    elif order in dict.keys() :
        tot+=dict[order]
        print(order+" is "+str(dict[order])+", total is",tot)
    else:
        print("Sorry, we are fresh out of "+order+" today!")
````

# Exercise 8: Biggest and smallest

1. Write a function, `biggest_and_smallest`, that takes a list of numbers.
2. The result from the function will be a two-element tuple containing the smallest number in the input list and the largest number in the input list.
3. Do *not* use any built-in Python functions for sorting, including `min` and `max`.
4. wouldn't it be nice if we could pass our numbers to the function *without* using a list? For this, we have *args

Example:

    biggest_and_smallest([53, 22, 97, 102, 15, 17, 12])
    
That will return the tuple

    (12, 102)

Hints:

- You're going to have to go through each element of the list
- Is it bigger than the biggest thing you've seen so far?  Or is it smaller than the smallest thing you've seen so far?

````python
def biggest_and_smallest(l):
    max=l[0]
    min=l[0]
    for i in l[1:]:
        if i>max:
            max=i
        if i<min:
            min=i
    return (min,max)
````

````python
biggest_and_smallest([53, 22, 97, 102, 15, 17, 12])
````
